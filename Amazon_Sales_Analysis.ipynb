## Group By
#### The groupby()  function in pandas is used to group data based on one or more columns in a DataFrame

```python
df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty', ascending=False)
```

```python
# Grouping, summing, and sorting
s_qty = df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty', ascending=False)

# Creating the bar plot with specified figure size
sns.set(style="whitegrid")  # Set the style

# Creating the bar plot
sns.barplot(x='Size', y='Qty', data=s_qty, palette='viridis')  # Adjust the palette for colors

# Adding labels and title
plt.xlabel('Size', fontsize=14)  # X-axis label with fontsize
plt.ylabel('Total Quantity', fontsize=14)  # Y-axis label with fontsize
plt.title('Total Quantity by Size', fontsize=16)  # Title with fontsize

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)  

# Show plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style
sns.set(style="whitegrid")

# Create the countplot
plt.figure(figsize=(8, 5))  # Adjust the figure size
sns.countplot(data=df, x='Courier Status', hue='Status', palette='Set2')  # Adjust palette for colors

# Add labels and title
plt.xlabel('Courier Status', fontsize=14)  # X-axis label with fontsize
plt.ylabel('Count', fontsize=14)  # Y-axis label with fontsize
plt.title('Courier Status Count by Status', fontsize=16)  # Title with fontsize

# Customize legend
plt.legend(title='Status', title_fontsize='14')

# Show plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style
sns.set(style="whitegrid")

# Create the countplot
plt.figure(figsize=(8, 5))  # Adjust the figure size
ax = sns.countplot(x='Category', data=df, palette='Set2')  # Adjust palette for colors

# Add labels and title
plt.xlabel('Category', fontsize=14)  # X-axis label with fontsize
plt.ylabel('Count', fontsize=14)  # Y-axis label with fontsize
plt.title('Count of Items by Category', fontsize=16)  # Title with fontsize

# Rotate x-axis labels for better readability
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

# Show plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

###                          Note: From above Graph you can see that most of the buyers are T-shirt


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Count the occurrences of each category in the 'B2B' column
b2b_counts = df['B2B'].value_counts()

# Plotting as a pie chart
plt.figure(figsize=(6,3))  # Adjust the figure size
plt.pie(b2b_counts, labels=b2b_counts.index, autopct='%1.1f%%', startangle=140)  # Plotting the pie chart
plt.title('Distribution of B2B', fontsize=16)  # Title with fontsize
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Show plot
plt.show()
```

### Note : From above chart we can see that maximum i.e. 99.2% of buyers are retailers and 0.8% are B2B buyers


```python
# Create a scatter plot
plt.figure(figsize=(8, 6))  # Adjust the figure size
sns.scatterplot(x='Category', y='Size', data=df, palette='viridis', s=100, alpha=0.7)  # Adjust palette for colors, marker size, and transparency

# Add labels and title
plt.xlabel('Category', fontsize=14)  # X-axis label with fontsize
plt.ylabel('Size', fontsize=14)  # Y-axis label with fontsize
plt.title('Scatter Plot of Size vs Category', fontsize=16)  # Title with fontsize

# Customize ticks
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Show plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Get the top 10 states with highest shipping frequencies
top_10_state = df['ship-state'].value_counts().head(10)

# Create a horizontal bar plot
plt.figure(figsize=(10, 6))  # Adjust the figure size
sns.barplot(x=top_10_state.values, y=top_10_state.index, palette='viridis')  # Horizontal bar plot with the 'viridis' palette

# Add labels and title
plt.xlabel('Frequency', fontsize=14)  # X-axis label with fontsize
plt.ylabel('State', fontsize=14)  # Y-axis label with fontsize
plt.title('Top 10 States by Shipping Frequency', fontsize=16)  # Title with fontsize

# Show plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

### Note: From above Graph you can see that most of the buyers are Maharashtra state
